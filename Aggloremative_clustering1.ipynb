```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering


data = np.array([
    [17, 15, 95, 43, 48],  
    [44, 81, 69, 86, 95],   
    [11, 18, 56, 13, 66],   
    [26, 31, 22, 50, 81],   
    [93, 18, 50, 55,100],   
    [18, 13, 40, 77, 84],   
    [11, 52, 49, 17, 79],   
    [93, 91, 71, 33, 29],   
    [80, 54, 49, 90, 12],   
    [77, 38, 67, 86,100],   
])

labels = ['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10']
```

```python
linked = linkage(data, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked, labels=labels, orientation='top', distance_sort='ascending')
plt.title("Dendrogram - Agglomerative Clustering (Ward)")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.grid(False)
plt.show()
```

```python
cluster_model = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters = cluster_model.fit_predict(data)

df = pd.DataFrame(data, index=labels, columns=["x1", "x2", "x3", "x4", "x5"])
df["Cluster"] = clusters
print(df)
```
