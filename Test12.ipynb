```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('application_train.csv')


df_filtered = df[df['DAYS_EMPLOYED'] < 0]


df_filtered['YEARS_EMPLOYED'] = -df_filtered['DAYS_EMPLOYED'] / 365


df_filtered['EMPLOYMENT_BIN'] = pd.cut(
    df_filtered['YEARS_EMPLOYED'],
    bins=[0, 1, 3, 5, 10, 20, 50],
    labels=['<1 yr', '13 yrs', '35 yrs', '510 yrs', '1020 yrs', '20+ yrs']
)


avg_credit_by_employment = df_filtered.groupby('EMPLOYMENT_BIN')['AMT_CREDIT'].mean().round(0)


plt.figure(figsize=(10, 5))
avg_credit_by_employment.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Loan Amount by Employment Duration')
plt.xlabel('Years Employed')
plt.ylabel('Average Loan Amount (AMT_CREDIT)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('application_train.csv')


df_clean = df[['REGION_POPULATION_RELATIVE', 'TARGET']].dropna()


correlation = df_clean['REGION_POPULATION_RELATIVE'].corr(df_clean['TARGET'])
print(f"Correlation between region population density and default rate: {correlation:.3f}")

plt.figure(figsize=(10, 6))
sns.regplot(
    x='REGION_POPULATION_RELATIVE',
    y='TARGET',
    data=df_clean,
    scatter_kws={'alpha': 0.2},
    line_kws={'color': 'red'}
)

plt.title('Default Rate vs. Region Population Density')
plt.xlabel('Region Population Relative (Proxy for Density)')
plt.ylabel('Default Rate (TARGET)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('application_train.csv')


working_df = df[df['NAME_INCOME_TYPE'] == 'Working']

edu_default = working_df.groupby('NAME_EDUCATION_TYPE')['TARGET'].mean().sort_values(ascending=False) * 100
edu_default = edu_default.round(2)

plt.figure(figsize=(10, 5))
edu_default.plot(kind='barh', color='steelblue', edgecolor='black')

plt.title('Default Rate by Education Level (Working Applicants Only)')
plt.xlabel('Default Rate (%)')
plt.ylabel('Education Level')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('application_train.csv')

df_clean = df[['AMT_ANNUITY', 'AMT_INCOME_TOTAL', 'NAME_HOUSING_TYPE']].dropna()


df_clean['ANNUITY_INCOME_RATIO'] = df_clean['AMT_ANNUITY'] / df_clean['AMT_INCOME_TOTAL']

housing_ratio = df_clean.groupby('NAME_HOUSING_TYPE')['ANNUITY_INCOME_RATIO'].mean().sort_values(ascending=False)
housing_ratio = (housing_ratio * 100).round(2)  

plt.figure(figsize=(10, 6))
housing_ratio.plot(kind='bar', color='mediumseagreen', edgecolor='black')

plt.title('Average Annuity-to-Income Ratio by Housing Type')
plt.xlabel('Housing Type')
plt.ylabel('Average Ratio (% of Income)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('application_train.csv')


df_clean = df[['AMT_ANNUITY', 'AMT_CREDIT']].dropna()
df_clean = df_clean[df_clean['AMT_CREDIT'] != 0]  
df_clean['ANNUITY_CREDIT_RATIO'] = df_clean['AMT_ANNUITY'] / df_clean['AMT_CREDIT']


df_clean = df_clean[df_clean['ANNUITY_CREDIT_RATIO'] < df_clean['ANNUITY_CREDIT_RATIO'].quantile(0.99)]


plt.figure(figsize=(10, 6))
sns.histplot(df_clean['ANNUITY_CREDIT_RATIO'], bins=50, kde=True, color='coral')

plt.title('Distribution of Annuity to Credit Ratio')
plt.xlabel('Annuity to Credit Ratio')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('application_train.csv')


df_clean = df[['EXT_SOURCE_1', 'NAME_CONTRACT_TYPE', 'TARGET']].dropna()


df_clean['EXT_BIN'] = pd.cut(df_clean['EXT_SOURCE_1'], bins=10)


grouped = df_clean.groupby(['EXT_BIN', 'NAME_CONTRACT_TYPE'])['TARGET'].mean().reset_index()


grouped['EXT_BIN_LABEL'] = grouped['EXT_BIN'].astype(str)


plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped, x='EXT_BIN_LABEL', y='TARGET', hue='NAME_CONTRACT_TYPE', marker='o')

plt.title('Default Rate by EXT_SOURCE_1 Score Bins and Loan Type')
plt.xlabel('EXT_SOURCE_1 Bins')
plt.ylabel('Default Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('application_train.csv')


org_default = df.groupby('ORGANIZATION_TYPE')['TARGET'].mean().sort_values(ascending=False)


highest_risk_org = org_default.idxmax()
print(f"Highest-risk organization type: {highest_risk_org}")


high_risk_df = df[df['ORGANIZATION_TYPE'] == highest_risk_org].copy()


high_risk_df['AGE'] = (-high_risk_df['DAYS_BIRTH'] / 365).round(1)


plt.figure(figsize=(10, 5))
sns.histplot(high_risk_df['AGE'], bins=30, kde=True, color='tomato')
plt.title(f'Age Distribution - {highest_risk_org}')
plt.xlabel('Age (Years)')
plt.ylabel('Applicant Count')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(high_risk_df['AMT_INCOME_TOTAL'], bins=40, kde=True, color='seagreen')
plt.title(f'Income Distribution - {highest_risk_org}')
plt.xlabel('Annual Income')
plt.ylabel('Applicant Count')
plt.grid(True)
plt.tight_layout()
plt.show()


sns.jointplot(
    data=high_risk_df,
    x='AGE',
    y='AMT_INCOME_TOTAL',
    hue='TARGET',
    kind='scatter',
    palette='coolwarm',
    alpha=0.6,
    height=8)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('application_train.csv')


df_clean = df[['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'TARGET']].dropna()


df_clean['Loan_Status'] = df_clean['TARGET'].map({0: 'Repaid', 1: 'Defaulted'})


grouped = df_clean.groupby(['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'Loan_Status']).size().reset_index(name='Count')


pivot_df = grouped.pivot_table(index=['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE'],
                                columns='Loan_Status',
                                values='Count',
                                fill_value=0)


pivot_df.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='coolwarm')

plt.title("Income Type + Education Level vs Loan Outcome")
plt.ylabel("Number of Applicants")
plt.xlabel("Income Type + Education Level")
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
